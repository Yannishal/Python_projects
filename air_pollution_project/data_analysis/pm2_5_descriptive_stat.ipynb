## Object variables to numeric

```python
df.pm2_5 = pd.to_numeric(df.pm2_5)
df.gdp_per_capita = pd.to_numeric(df.gdp_per_capita)
df.access_to_e_per = pd.to_numeric(df.access_to_e_per)
df.urban_pop = pd.to_numeric(df.urban_pop)
```

```python
df.info()
```

```python
df['urban_pop'] = df['urban_pop']/1000000 # turn population to millions
```

## Boxplots

```python
fig = go.Figure()
```

```python
fig.add_trace(go.Box(y = df['pm2_5'], name = 'PM2.5 Boxplot',
                     boxpoints='all',
            jitter=0.5,
            width = 0.1,
            marker_color='#FF851B',
            hovertext = df['country'],
            whiskerwidth=0.2,
            marker_size=1.5,
            line_width=1))
```

```python
fig_2 = go.Figure()
```

```python
fig_2.add_trace(go.Box(y = df['access_to_e_per'], name = 'Electricity Access Boxplot',
                     boxpoints='all',
            jitter=0.5,
            width = 0.1,
            marker_color='#FF851B',
            hovertext = df['country'],
            whiskerwidth=0.2,
            marker_size=1.5,
            line_width=1)),
fig_2.update_layout( yaxis=dict(
        autorange=True,
        showgrid=True,
        zeroline=True,
        dtick=5,
        gridcolor='rgb(255, 255, 255)',
        gridwidth=1,
        zerolinecolor='rgb(255, 255, 255)',
        zerolinewidth=2,
    ),
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0,
    ),
    paper_bgcolor='rgb(243, 243, 243)',
    plot_bgcolor='rgb(243, 243, 243)',
    showlegend=False
)

fig_2.show()
```
