# Data collection

```python
import pandas as pd     #world bank data library for faster access
import numpy as np
import wbgapi as wb
```

```python
df = wb.data.DataFrame(['EN.ATM.PM25.MC.M3', 'NY.GDP.PCAP.CD', 'EG.ELC.ACCS.ZS', 'SP.URB.TOTL'], skipAggs=True, time=range(2010,2018), labels=True)
```

```python
df = df.reset_index()
```

```python
df
```

# Data Cleaning 

```python
df.head()
```

```python
df.drop(['series'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
columns = ['code', 'country', 'series', '2010', '2011', '2012', '2013', '2014',
          '2015', '2016', '2017']
```

```python
df.columns = columns
```

```python
df.head() #new names 
```

```python
df.series.unique()
```

```python
df_pm2_5 = df[df['series'] == 'PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)']                #create df of each variable to transfrom data from wide to long type
df_gdppc = df[df['series'] == 'GDP per capita (current US$)']
df_access_to_ele = df[df['series'] == 'Access to electricity (% of population)']
df_urban_pop = df[df['series'] == 'Urban population']
```

```python
df_pm2_5 = df_pm2_5.set_index(['code', 'country', 'series']).stack().reset_index()
df_gdppc = df_gdppc.set_index(['code', 'country', 'series']).stack().reset_index()
df_access_ele = df_access_to_ele.set_index(['code', 'country', 'series']).stack().reset_index()
df_urban_pop = df_urban_pop.set_index(['code', 'country', 'series']).stack().reset_index()
```

```python
columns_pm2_5 = ['code', 'country', 'series', 'year', 'pm2_5']
columns_gdppc = ['code', 'country', 'series', 'year', 'gdp_per_capita']
columns_access_to_ele = ['code', 'country', 'series', 'year', 'access_to_e_per']
columns_urban = ['code', 'country', 'series', 'year', 'urban_pop']
```

```python
df_pm2_5.columns = columns_pm2_5
df_gdppc.columns = columns_gdppc
df_access_ele.columns = columns_access_to_ele
df_urban_pop.columns = columns_urban
```

# Merging the Dataframes

```python
df_pm2_5.info() #transform all floats to object type in order to merge the dataset 
```

```python
df_pm2_5['pm2_5'] = df_pm2_5['pm2_5'].astype(str)
```

```python
df_gdppc['gdp_per_capita'] = df_gdppc['gdp_per_capita'].astype(str)
```

```python
df_access_ele['access_to_e_per'] = df_access_ele['access_to_e_per'].astype(str)
```

```python
df_urban_pop['urban_pop'] = df_urban_pop['urban_pop'].astype(str)
```

```python
df_pm2_5.drop(['series'], axis = 1, inplace = True) # drop the series in order to merge
```

```python
df_gdppc.drop(['series'], axis = 1, inplace = True)
```

```python
df_access_ele.drop(['series'], axis = 1, inplace = True)
```

```python
df_urban_pop.drop(['series'], axis = 1, inplace = True)
```

```python
df_1 = pd.merge(df_pm2_5, df_gdppc)
```

```python
df_1
```

```python
df_2 = pd.merge(df_1, df_access_ele)
```

```python
df_2
```

```python
final_df = pd.merge(df_2, df_urban_pop)
```

```python
final_df
```

## Remove null values from final dataframe

```python
final_df.drop(final_df[(final_df['pm2_5'] == '..')].index, inplace = True)
```

```python
final_df.drop(final_df[(final_df['gdp_per_capita'] == '..')].index, inplace = True)
```

```python
final_df.drop(final_df[(final_df['access_to_e_per'] == '..')].index, inplace = True)
```

```python
final_df.drop(final_df[(final_df['urban_pop'] == '..')].index, inplace = True)
```
